```python
# !pip install pandas pyarrow
import pandas as pd
```

```python
df1 = pd.read_csv('dados/venda_passagem_12_2019.csv',
                  sep=';',
                  encoding='latin1',
                  decimal=',',
                  thousands='.',
                  on_bad_lines='skip'
)
```

```python
df2 = pd.read_csv('dados/venda_passagem_11_2019.csv',
                  sep=';',
                  encoding='latin1',
                  decimal=',',
                  thousands='.',
                  on_bad_lines='skip'
)
```

```python
df3 = pd.read_csv('dados/venda_passagem_03_2019.csv',
                  sep=';',
                  encoding='latin1',
                  decimal=',',
                  thousands='.',
                  on_bad_lines='skip'
)
```

```python
df4 = pd.read_csv('dados/venda_passagem_01_2020.csv',
                  sep=';',
                  encoding='latin1',
                  decimal=',',
                  thousands='.',
                  on_bad_lines='skip'
)
```

```python
df_final = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
# Conversão de compos VALOR
df_final['valor_tarifa'] = df_final['valor_tarifa'].astype(float)
df_final['valor_percentual_desconto'] = df_final['valor_percentual_desconto'].astype(float)
df_final['valor_aliquota_icms'] = df_final['valor_aliquota_icms'].astype(float)
df_final['valor_pedagio'] = df_final['valor_pedagio'].astype(float)
df_final['valor_taxa_embarque'] = df_final['valor_taxa_embarque'].astype(float)
df_final['valor_total'] = df_final['valor_total'].astype(float)
```

```python
# Conversão de compos DATA e HORA
df_final['data_emissao_bilhete'] = df_final['data_emissao_bilhete'].astype('datetime64[ns]')
df_final['data_viagem'] = df_final['data_viagem'].astype('datetime64[ns]')
# df_final['data_viagem'] = pd.to_datetime(df_final['data_viagem'], format='%Y-%m-%d')
# df_final['data_emissao_bilhete'] = pd.to_datetime(df_final['data_emissao_bilhete'], format='%Y-%m-%d')
# df_final['data_emissao_bilhete'] = pd.to_datetime(df_final['data_emissao_bilhete'], format='%Y-%m-%d')
# f_final['data_viagem'] = pd.to_datetime(df_final['data_viagem'], format='%Y-%m-%d')
```
